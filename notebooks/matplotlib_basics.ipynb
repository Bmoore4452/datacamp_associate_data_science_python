```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)
plt.show()
```

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.scatter(year, pop)
plt.show()
```

```python
import random

random_integers = [random.randint(0, 9) for _ in range(16)]

plt.hist(random_integers, bins=3, edgecolor='black')

plt.show()
```

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
ticks = [0, 2, 4, 6, 8, 10]
tick_labels = ['0', '2B', '4B', '6B', '8B', '10B']

year = [1850, 1900, 1950] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year, pop)

plt.xlabel("Year")
plt.ylabel("Population")
plt.title("World Population Projections")
plt.yticks(ticks, tick_labels)

plt.show()
```
